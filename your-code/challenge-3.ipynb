## Q1: How to identify VIP & Preferred Customers?

Next, extract and import `Orders` dataset into a dataframe variable called `orders`. Print the head of `orders` to overview the data:

## How to label customers whose aggregated `amount_spent` is in a given quantile range?


We break down the main problem into several sub problems:

#### Sub Problem 1: How to aggregate the  `amount_spent` for unique customers?

#### Sub Problem 2: How to select customers whose aggregated `amount_spent` is in a given quantile range?

#### Sub Problem 3: How to label selected customers as "VIP" or "Preferred"?

*Note: If you want to break down the main problem in a different way, please feel free to revise the sub problems above.*

Now in the workspace below, tackle each of the sub problems using the iterative problem solving workflow. Insert cells as necessary to write your codes and explain your steps.

```python
# Sub Problem 1: How to aggregate the  `amount_spent` for unique customers?

customers_with_amount_spent = orders.groupby('CustomerID').agg({'amount_spent':sum}).reset_index()
customers_with_amount_spent.sort_values(by='amount_spent', ascending=False)
```

```python
quantile95 = customers_with_amount_spent.quantile(0.95)
print(quantile95)

print('========================')

quantile75 = customers_with_amount_spent.quantile(0.75)
print(quantile75)
```

```python
# Sub Problem 2: How to select customers whose aggregated `amount_spent` is in a given quantile range?

# Quantile 95
vip_customers = customers_with_amount_spent[customers_with_amount_spent['amount_spent'] > quantile95.amount_spent]
vip_customers.sort_values(by='amount_spent', ascending=False)
```

```python
# Quantile 75
preferred_customer = customers_with_amount_spent[(customers_with_amount_spent['amount_spent'] > quantile75.amount_spent) & (customers_with_amount_spent['amount_spent'] < quantile95.amount_spent)]
preferred_customer.sort_values(by='amount_spent', ascending=False)
```

## Q2: How to identify which country has the most VIP Customers?

```python
customer_by_country = orders.groupby(['Country', 'CustomerID']).agg({'amount_spent':sum}).reset_index()
customer_by_country
```

```python
vip_customer_by_country = customer_by_country[customer_by_country['amount_spent'] > quantile95.amount_spent]
vip_customer_by_country.groupby(['Country']).agg({'amount_spent':len}).sort_values(by='amount_spent', ascending=False)
```

## Q3: How to identify which country has the most VIP+Preferred Customers combined?

```python
vip_preferred_customer_by_country = customer_by_country[customer_by_country['amount_spent'] > quantile75.amount_spent]
vip_customer_by_country.groupby(['Country']).agg({'amount_spent':len}).sort_values(by='amount_spent', ascending=False)
```
